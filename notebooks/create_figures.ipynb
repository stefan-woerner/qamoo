```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
figures_folder = './../figures/'
```

```python
x_opt = np.array([0.05, 0.35, 0.5, 0.85, 1])
y_opt = np.array([1, 0.85, 0.5, 0.35, 0.05])

ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

eps = 0.01
plt.plot(x_opt[[0,1,3,4]]+eps, y_opt[[0,1,3,4]]+eps, '--', color='blue')
plt.plot(x_opt, y_opt, 'o-', color='green')

plt.fill_between([0, x_opt[0]], [0, 0], [y_opt[0], y_opt[0]], color='green', alpha=0.25, edgecolor='none', linewidth=0)
for i in range(1, 5):
    plt.fill_between([x_opt[i-1], x_opt[i]], [0, 0], [y_opt[i], y_opt[i]], color='green', alpha=0.25, edgecolor='none', linewidth=0)

f = 0.8
x_approx = f*np.array([0, 0.2, 0.4, 0.6, 0.8])
y_approx = f*np.array([0.8, 0.6, 0.4, 0.2])

for i in range(1, 5):
    plt.fill_between([x_approx[i-1], x_approx[i]], [0, 0], [y_approx[i-1], y_approx[i-1]], color='orange', alpha=0.25, edgecolor='none', linewidth=0)
    plt.plot(x_approx[i], y_approx[i-1], 'D', color='orange')

color = 'orange'

plt.plot([0, 0], [0, f*0.8], '--', color=color)
plt.plot([f*0.2, f*0.2], [0, f*0.8], '--', color=color)
plt.plot([f*0.4, f*0.4], [0, f*0.6], '--', color=color)
plt.plot([f*0.6, f*0.6], [0, f*0.4], '--', color=color)
plt.plot([f*0.8, f*0.8], [0, f*0.2], '--', color=color)

plt.plot([0, f*0.2], [f*0.8, f*0.8], '--', color=color)
plt.plot([0, f*0.4], [f*0.6, f*0.6], '--', color=color)
plt.plot([0, f*0.6], [f*0.4, f*0.4], '--', color=color)
plt.plot([0, f*0.8], [f*0.2, f*0.2], '--', color=color)
plt.plot([0, f*0.8], [0.0, 0.0], '--', color=color)

plt.plot(0, 0, 'rD')

plt.xlabel(r'$f_1$', fontsize=14)
plt.ylabel(r'$f_2$', fontsize=14)
plt.tight_layout()
plt.savefig(f'{figures_folder}hyper_volume.pdf')
plt.show()
```

```python
x_opt = np.array([0, 0.35, 0.5, 0.85, 1])
y_opt = np.array([1, 0.85, 0.5, 0.35, 0])

ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

delta = 0.03
slope = (0.5-0.85)/(0.5-0.35)

plt.hlines(0.35 + delta, 0, 1, colors='gray', linestyles='--')
plt.vlines(0.35 + delta, 0, 1, colors='gray', linestyles='--')

plt.fill_between([0.35+delta, 0.5, 0.85+slope*delta], delta + np.array([0.35, 0.35, 0.35]), [0.85+slope*delta, 0.5, 0.35+delta], color='green', alpha=0.25, edgecolor='none', linewidth=0)

eps = 0.01
plt.plot(x_opt[[0,1,3,4]]+eps, y_opt[[0,1,3,4]]+eps, '--', color='blue')
plt.plot(x_opt, y_opt, 'o-', color='green')

plt.text(0.4, 0.0, r'$\epsilon_1$')
plt.text(0.0, 0.4, r'$\epsilon_2$')
plt.xlabel(r'$f_1$', fontsize=14)
plt.ylabel(r'$f_2$', fontsize=14)
plt.tight_layout()
plt.savefig(f'{figures_folder}epsilon_constraints.pdf')
plt.show()
```
